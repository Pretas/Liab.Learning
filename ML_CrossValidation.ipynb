```
!pip install pymysql
!pip install pymssql

import pandas as pd
import numpy as np
import LibData
import LibUtils

#jobName = '201216_Train_pilot'
jobName = '201216_Train_split1'
x, y = LibData.GetXY3(jobName)

# 스케일링 없이
cols_toOneHot = ['ProdCode']
x = pd.get_dummies(x, columns=cols_toOneHot, prefix='prd')

from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR
from sklearn import metrics

model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)

# cv 실행
scores = cross_val_score(model, x, y['pv_claim'], cv=5, scoring = 'neg_mean_absolute_error')

LibUtils.WriteNote('T2_scaler(svm)_data10_mae', 'no scaler', np.append('avg'+str(np.mean(scores)), scores))

LibUtils.ELoop()
```
