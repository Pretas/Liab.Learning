```python
import pandas as pd
import DataLoad
import LibUtils

jobName = '201216_Train'
x, y = DataLoad.GetXY4(jobName, 300000)

# 범주형 변수 원핫인코딩
x = pd.get_dummies(x, columns=['ProdCode'], prefix='prd')

print(x.columns, y.columns)
print(len(x), len(y))
```

```python
import multiprocessing # 병렬 처리를 위한 모듈

multiprocessing.cpu_count() # 8개 코어
```

```python
################################################
# 그리드 서치
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

tuned_parameters = {
    'n_estimators' : [100, 500],
    'max_depth' : [5, 15],
    'min_samples_split' : [10],
    'min_samples_leaf' : [3]
}

scores = 'neg_mean_squared_error'

gridSearcher = GridSearchCV(RandomForestRegressor(), param_grid = tuned_parameters, cv=5, scoring=scores, n_jobs=multiprocessing.cpu_count(), verbose=1)
gridSearcher.fit(x, y['pv_claim'])
```

```python
LibUtils.WriteNote('bets param rf', gridSearcher.best_params_)
```
