```
# 트레이닝, 테스트 데이터 생성
xTrain, yTrain, embs, scaler = LibData.TrainingDataFinal('201216_Train_split1')
xTest, yTest = LibData.TestDataFinal('201216_Test_TrainScen_split20000', embs, scaler)
```

```
# XGBoost
md = LibML.XGBoostTuned()
pred = LibML.GetPredBase(xTrain, yTrain, xTest, md)
print(LibML.GetScoresOV(yTest, pred))
```

```
#SVM
from sklearn.svm import SVR

md = SVR(kernel = 'rbf', C = 184, gamma = 0.162)
pred = LibML.GetPredBase(xTrain, yTrain, xTest, md)
print(LibML.GetScoresOV(yTest, pred))
```

```
# RF
md = LibML.RandomForestTuned(len(xTrain.columns))
pred = LibML.GetPredBase(xTrain, yTrain, xTest, md)
print(LibML.GetScoresOV(yTest, pred))
```

```
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso

xTrain_poly = pd.DataFrame(PolynomialFeatures(2).fit_transform(xTrain))
xTest_poly = pd.DataFrame(PolynomialFeatures(2).fit_transform(xTest))

# Ridge
ridge = Ridge(alpha = 0.147)
pred = LibML.GetPredBase(xTrain_poly, yTrain, xTest_poly, ridge)
print(LibML.GetScoresOV(yTest, pred))

# Lasso
lasso = Lasso(alpha = 0.185)
pred = LibML.GetPredBase(xTrain_poly, yTrain, xTest_poly, lasso)
print(LibML.GetScoresOV(yTest, pred))
```
